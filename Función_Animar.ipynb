```python
import numpy as np

import matplotlib.pyplot as plt

import matplotlib.animation as animation
```

```python
## Para animar la Densidad. 

def AnimacionDens(xi, dens, Nombre, frames, fps, xmin, xmax): 
    
      
    """ frames ==> Numero de capturas totales. 
        fps ==> Numero de por segundo. 
        Nombre ==> Nombre para nuestro GIFT 
        xi ==> Puntos dominio de la funci贸n u(x). 
        u ==> funci贸n u(x) para cada t. 
    """

    fig, ax = plt.subplots(figsize=(4,4))

    line2 = ax.plot(xi, dens[0])[0]


    def update(frame):
        # for each frame, update the data stored on each artist.
        x = xi
        y = dens[frame]
        # update the scatter plot:
        line2.set_xdata(x)
        line2.set_ydata(y)
        return line2


    ani = animation.FuncAnimation(fig=fig, func=update, frames = frames)
    ax.set_title(Nombre)
    ax.set_xlim(xmin,xmax)
    #ax.set_ylim(max(u[0]),min(u[-1]))
    ax.set_ylabel(r'$\rho(x,t)$')
    ax.set_xlabel(r'$x$')
    plt.show()
    ani.save(Nombre, writer='PillowWriter', fps = fps)
    
## Para animar el perfil. 
    
def Animacion(xi, u, Nombre, frames, fps): 
    
    """ frames ==> Numero de capturas totales. 
        fps ==> Numero de por segundo. 
        Nombre ==> Nombre para nuestro GIFT 
        xi ==> Puntos dominio de la funci贸n u(x). 
        u ==> funci贸n u(x) para cada t. 
    """
        

    fig, ax = plt.subplots(figsize=(4,4))

    line2 = ax.plot(xi, u[0])[0]


    def update(frame):
        # for each frame, update the data stored on each artist.
        x = xi
        y = u[frame]
        # update the scatter plot:
        line2.set_xdata(x)
        line2.set_ydata(y)
        return line2


    ani = animation.FuncAnimation(fig=fig, func=update, frames = frames)
    ax.set_xlim(xi[0],xi[-1])
    #ax.set_ylim(max(u[0]),min(u[-1]))
    ax.set_title(Nombre)
    ax.set_ylabel(r'$u(x,t)$')
    ax.set_xlabel(r'$x$')
    plt.show()
    # Comentar si no se qquiere guardar el gift. 
    ani.save(Nombre, writer='PillowWriter', fps = fps)
    
```
